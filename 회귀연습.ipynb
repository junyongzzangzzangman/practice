```python
import pandas as pd
import numpy as np
X_train=pd.read_csv('t2-2-X_train.csv')
y_train=pd.read_csv('t2-2-y_train.csv')
X_test=pd.read_csv('t2-2-X_test.csv')
X_train=X_train.iloc[:,1:8]
X_test2=X_test.iloc[:,1:8]

from sklearn.model_selection import train_test_split
X,X_val,y,y_val=train_test_split(X_train,y_train)


from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
model=RandomForestRegressor()
model.fit(X,y['Chance of Admit '])
pred_train=model.predict(X)
pred_val=model.predict(X_val)

from sklearn.metrics import mean_squared_error
MSE_train=mean_squared_error(y['Chance of Admit '],pred_train)
MSE_val=mean_squared_error(y_val['Chance of Admit '],pred_val)
print("훈련 RMSE:",np.sqrt(MSE_train))
print("검증 RMSE:",np.sqrt(MSE_val))

pred_test=model.predict(X_test2)
pred_test
sub=pd.concat([pd.DataFrame(X_test[['Serial No.']]),pd.DataFrame(pred_test)],axis=1)
sub.columns=['Serial No.','Chance of Admit ']
sub.to_csv("20220624.csv",index=False)
pd.read_csv("20220624.csv")
```
